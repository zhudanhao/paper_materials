```python
from rouge_chinese import Rouge
import jieba
import json
```

```python
# 获取待评价的数据,数据以[{"y":xxx, "pred":xxx}]形式保存在json文件中
# predict_y_gpt3.5_200.json
file = open("Qwen-main/data/eval_data/JI_FEW_0628_V2340.json",'r', encoding='utf-8')
content = file.read()
eval_data = json.loads(content)
# eval_data = eval_data[:100]
```

```python
for i in range(len(eval_data)):
    if len(eval_data[i]["pred"])==0:
        eval_data[i]["pred"] = "本次生成失败"
```

```python
# 将数据格式化到列表中
hyps = []
refs = []
for data in eval_data:
    hyps.append(' '.join(jieba.cut(data["y"])))
    refs.append(' '.join(jieba.cut(data["pred"])))
```

```python
# 进行测试，并将测试结果进行输出
rouge = Rouge()
scores = rouge.get_scores(hyps, refs, avg=True)
```
